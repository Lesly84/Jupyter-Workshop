```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2*np.pi*t)
plt.plot(t, s)

plt.xlabel('Time (s)')
plt.ylabel('Swag')
plt.title('Wiggle Wiggle')
plt.grid(True)
plt.savefig("test.png")
plt.show()
```
